```python
import numpy as np
x = np.linspace(-2,2,500)

y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
# afegm titol
plt.title("Square function")
# labels
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

plt.plot(x, y, "b--",x, y2,"g")
plt.show()
```

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

plt.plot(x, y, "b--")
plt.plot(x, y2,"g")
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

plt.plot(x, y, "b--", label = "x**2")
plt.plot(x, y2,"g", label = "x + 1")
plt.legend(loc = "best")
plt.show()
```

```python
# dos figures que no es superposin
x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

#(files, columnes, posició)
plt.subplot(1,2,1)
plt.plot(x, y, "b--")
plt.subplot(1,2,2)
plt.plot(x, y2,"g")
plt.show()
```

```python

plt.figure(figsize=(14,6))
x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

#(files, columnes, posició)
plt.subplot(1,2,1)
plt.plot(x, y, "b--")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y = x**2", fontsize = 14)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x, y2,"g")
plt.xlabel("x2", fontsize = 14)
plt.ylabel("y2 = y +1 ", fontsize = 14)
plt.grid(True)
plt.show()
```

# Scatter Plots


```python
from numpy.random import rand
x,y = rand(2,100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x2,y2 = rand(2,100)
plt.scatter(x,y, c = "red")
plt.scatter(x2,y2, c = "blue")

plt.show()
```

 # Histogramas
 

```python
data = [1,1.1,1.8,2,2.1,3.2,3,3,3,3,3]
# dos files, 1 columa i primera posició però mostrat amb un unic valor
plt.subplot(211)
plt.hist(data, bins= 15, rwidth =0.8)
plt.show()
```

```python
# guardar figuras
import numpy as np
plt.figure(figsize=(14,6))

x = np.linspace(-2,2,500)
y = x**2
y2 = x +1

#(files, columnes, posició)
plt.subplot(1,2,1)
plt.plot(x, y, "b--")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y = x**2", fontsize = 14)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x, y2,"g")
plt.xlabel("x2", fontsize = 14)
plt.ylabel("y2 = y +1 ", fontsize = 14)
plt.grid(True)

plt.savefig("migrafica.png", transparent = True)
```
